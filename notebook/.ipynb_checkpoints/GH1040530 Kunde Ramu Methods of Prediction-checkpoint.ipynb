## Data Exploration

- Torchvision package is used to load Image data folder from dataset and transforms for image transformation for like resizaing, normalization 
- torch.Size(32, 3, 128, 128) where it indicates that batch size=32, no.of channels=3, height=128, width=128.

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tensorflow.keras.layers import Rescaling

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

dataset = datasets.ImageFolder(root='C:/Users/DELL/Downloads/dogcat', transform=transform)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

images, labels = next(iter(loader))
print(images.shape, labels)
```

- os module is used to interact with the operating system, specifically for navigating directories, matplotlib.pyplot module is used for plotting, commonly used for displaying images and tensorflow library, which is used here for loading images.
- os.listdir(train_dir) for listing label used in dataset
- imported two dataset folder for train and test and also Displayed five images from train folder of each Label.

```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf

train_dir = 'C:/Users/DELL/Downloads/dogcat/train'
test_dir = 'C:/Users/DELL/Downloads/dogcat/test'

print(os.listdir(train_dir))
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    for img_file in os.listdir(class_path)[:3]:
        img = tf.keras.utils.load_img(os.path.join(class_path, img_file))
        plt.imshow(img)
        plt.title(class_name)
        plt.show()
```

## Data Preprocessing & Feature Engineering

- Importing important libraries for further task performing sequential 

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
import pandas as pd
from tabulate import tabulate
```

- keras.utils.image_dataset_from_directory function is used to efficiently load image datasets from directories and prepare them for training CNN model
- train_ds is used for training data and validation_ds for test with labels=inferred for labels from the subdirectory, batch_size for bunch of 16 images and image _size of 64x64 pixels
- found 20000 images with 2 labels from train_ds and 5000 images with 2 labels from test_dir

```python
train_ds = keras.utils.image_dataset_from_directory(
    train_dir,
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 16,
    image_size=(64,64)
)

validation_ds = keras.utils.image_dataset_from_directory(
    test_dir,
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 16,
    image_size=(64,64)
)
```

- Normalizes the image pixel values to the range of 0 to 1 and converts them to a floating-point data type, which is a standard preprocessing step for training models on image data.

```python
def process(image, label):
  image = tf.cast(image/255., tf.float32)
  return image, label

train_ds = train_ds.map(process)
validation_ds = validation_ds.map(process)
```

## Model Design & Training
- Creating convolutional neural network model with sequential which adds one after another in sequence
-  First layer with 16 filters with size(3,3) and input_shape(64,64,3) it specifices height, width RGB channels respectively and second layer with 32 filters.
-  In this batchnormalization is used to normalizes the activations of the previous layers to stabilizes the learning process.
- Maxpooling is used to reduce height and width of the feature maps, which helps to reduce the no.of parameters and Flatten layer is used for converts the 2D feature maps into 1D vector.
- Dense layer are fully connected layers where each neuron is connected to every neuron in the previous layer. The activation function ReLU is used for hidden layers, and sigmoid is used for the output layer in binary classification to output a probability between 0 and 1.
- Drop out randomly set a fraction of the input units to 0.1 at each training update, which helps to prevent overfitting.

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(64,64,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())

model.add(Dense(64, activation='sigmoid'))
model.add(Dropout(0.1))
model.add(Dense(64, activation= 'sigmoid'))
model.add(Dropout(0.1))
model.add(Dense(1, activation= 'sigmoid'))    
```

## Model summary

- The model.summary() method is used to prints a summary of model's architecture it includes name and type of each layer in the model and shape of the output from each layer. None in batch size dimension shows that  model can accepts any batch size and the Total params: 207,105 (809.00 KB),Trainable params: 207,009 (808.63 KB),Non-trainable params: 96 (384.00 B).

```python
model.summary()
```

- Model.compile() method is used to configure model training and this model uses  optimizer=adam and loss=binary_crossentropy for classification of cat and dog so that the model can predict probability to match the true labels. Metrics 'accuracy' is used to evaluate the model's performance during training and testing.

```python
from tensorflow.keras.metrics import Precision, Recall, AUC
import numpy as np
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy',Precision(), Recall(), AUC()])
```

- The model fit method trains model for 10 epochs and the traning progress along performances metrics is stored in history variable.

```python
history = model.fit(train_ds,epochs=10,validation_data=validation_ds)
```

##  Visualization Precision: train vs validation & Visualization Recall: train vs validation

```python
df['optimizer'] = df['optimizer'].str.capitalize()
optimizers = df['optimizer'].unique()
fig, axs = plt.subplots(len(optimizers), 2, figsize=(8, 4 * len(optimizers)))
for i, opt in enumerate(optimizers):
    opt_data = df[df['optimizer'] == opt].iloc[0]

    # Precision pie chart
    axs[i, 0].pie(
        [opt_data['precision'], opt_data['val_precision']],
        labels=['Train Precision', 'Val Precision'],
        colors=['purple', 'grey'],
        autopct='%1.1f%%',
        startangle=90
    )
    axs[i, 0].set_title(f'{opt} - Precision')

    # Recall pie chart
    axs[i, 1].pie(
        [opt_data['recall'], opt_data['val_recall']],
        labels=['Train Recall', 'Val Recall'],
        colors=['green', 'yellow'],
        autopct='%1.1f%%',
        startangle=90
    )
    axs[i, 1].set_title(f'{opt} - Recall')

plt.tight_layout()
plt.show()
```

## Model Evaluation On Test set

- Now test a trained model using new images first scale the image pixels to 0-1 range for better processing then load test images from a folder,resizes them, and groups them into batches.After that we apply  scaling, the model evaluates its performance on these images, and the final loss and accuracy are printed

```python
normalization_layer = Rescaling(1./255)
test_ds = keras.utils.image_dataset_from_directory(test_dir, image_size=(64, 64), batch_size=16)
test_ds = test_ds.map(lambda x, y: (normalization_layer(x), y))

results = model.evaluate(test_ds)
print("Test Loss, Test Accuracy:", results)
```
